The goal is to understand relationships between clinical attributes and the likelihood of heart disease, forming the foundation for predictive modeling in the *Heart Disease Risk Predictor* project.

## Missing Values & Data Cleaning

Preliminary inspection revealed some missing entries, particularly in the `ca` (number of major vessels) and `thal` (thalassemia) columns.  
To maintain dataset usability without dropping valuable samples, simple imputation strategies were applied:

- **`ca`**: replaced missing values with the **mean** of the column, since it’s a numerical feature.  
- **`thal`**: filled missing values with the **mode**, as it represents categorical levels.

### Rationale
- Mean imputation preserves numeric continuity for regression-based models.  
- Mode imputation avoids creating artificial intermediate categories in `thal`.  
- This step retains all rows, preventing potential bias from record deletion.

After imputation, the dataset was rechecked using `.info()` to confirm no remaining null values and ensure consistent data types across features.


```python
df['target'] = df['target'].apply(lambda x: 1 if x > 0 else 0)
```

```python
mean = df['ca'].mean()
mode = df['thal'].mode()[0]
df['ca'] = df['ca'].fillna(mean)
df['thal'] = df['thal'].fillna(mode)
```

```python
df.info()
```

## Data Exploration & Visualization

Before visualizing feature relationships, the `target` column was simplified from a multiclass label (0–4) into a **binary classification**:  
`1` indicates **presence of heart disease**, while `0` indicates **no disease**.

### Why This Transformation?
- The original UCI dataset uses values 1–4 to represent different severity levels of heart disease.  
- For initial modeling and interpretability, it’s more practical to approach this as a **binary classification problem** — predicting whether a patient *has* or *doesn’t have* heart disease.  
- This simplification aligns with most published research and common ML benchmarks for this dataset.

---

### Visual Analysis
The following visualizations explore relationships and distributions in the dataset:

1. **Correlation Heatmap** – highlights linear relationships between numeric features.  
2. **Feature Distributions** – histograms for continuous features to identify skewness or outliers.  
3. **Categorical Counts** – count plots grouped by `target` to observe differences between patients with and without heart disease.  
4. **Mutual Information Scores** – quantifies how informative each feature is about the target.

These plots collectively reveal which features carry the strongest predictive potential and guide future preprocessing steps (like scaling or encoding).


```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

```python
features = [
    "age", "sex", "cp", "trestbps", "chol", "fbs", "restecg",
    "thalach", "exang", "oldpeak", "slope", "ca", "thal"]
```

```python
X_full = df[features]
y_full = df['target']
```

```python
from sklearn.feature_selection import mutual_info_classif
```

```python
mi_scores = mutual_info_classif(X_full, y_full, discrete_features='auto', random_state=42)
mi_df = pd.DataFrame({'Feature': X_full.columns, 'Mutual_Info': mi_scores})
mi_df = mi_df.sort_values(by='Mutual_Info', ascending=False)
mi_df
```

```python
mi_df = mi_df.sort_values(by='Mutual_Info', ascending=True)

plt.figure(figsize=(8, 6))
plt.barh(mi_df['Feature'], mi_df['Mutual_Info'], color='teal')
plt.title('Mutual Information Scores')
plt.xlabel('MI Score')
plt.ylabel('Feature')
plt.show()
```

## Insights & Next Steps

### Key Findings
- **Age**, **maximum heart rate (thalach)**, and **ST depression (oldpeak)** show strong relationships with heart disease presence.  
- **Chest pain type (cp)** and **exercise-induced angina (exang)** also display clear class separations between healthy and diseased patients.  
- **Serum cholesterol (chol)** and **resting blood pressure (trestbps)** show weaker correlations, suggesting they may be secondary indicators rather than primary predictors.  
- **Correlation heatmap** confirmed mild multicollinearity among some numerical features, which may later require scaling or dimensionality checks.  
- **Mutual information analysis** ranked `cp`, `thalach`, and `oldpeak` as top contributors, supporting their use in early feature selection.  
- Data imbalance is minimal — roughly equal representation of patients with and without heart disease, which reduces the need for resampling at this stage.  

### Data Quality Notes
- Missing values in `ca` and `thal` were imputed logically and verified post-cleaning.  
- No duplicate rows were detected, ensuring data integrity.  
- Outliers (e.g., unusually high cholesterol) were inspected but retained to preserve real-world variability.

### Next Steps
1. **Preprocessing Pipeline**
   - Apply encoding for categorical variables (`thal)`.  
   - Standardize continuous features using `StandardScaler`.  

2. **Model Development**
   - Begin with Logistic Regression as a baseline classifier.  
   - Experiment with tree-based methods (Decision Tree, Random Forest, XGBoost) to capture non-linear relationships.  

3. **Evaluation & Deployment**
   - Compare models using ROC–AUC, precision, recall, and F1-score.  
   - Save and deploy the final model in a Streamlit app for demonstration.  

This concludes the exploratory phase. The cleaned and analyzed dataset now provides a solid foundation for predictive modeling in the **Heart Disease Risk Predictor** project.
